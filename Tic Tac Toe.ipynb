```python
place = [] #create an empty list to store your values in your grid

for i in range (0, 9) : #going to go through i 9 times
    place.append(str(i + 1))
    
def grid():
    print( '  ┌─────┬─────┬─────┐')
    print( '  │  ' + place[0] + '  │  ' + place[1] + '  │  ' + place[2] + '   |  ')
    print( '  ├─────┼─────┼─────┤')
    print( '  │  ' + place[3] + '  │  ' + place[4] + '  │  ' + place[5] + '  |  ')
    print( '  ├─────┼─────┼─────┤')
    print( '  │  ' + place[6] + '  │  ' + place[7] + '  │  ' + place[8] + '  |  ')
    print( '  └─────┴─────┴─────┘')
p1 = True
win = False
while win == False:
    grid()
    
    if p1:
        print("It's Player Ones go!")
    else:
        print("It's Player Twos go!")
    
    try: 
        turn = int(input())
    except:
        print("Thats not allowed! Try again")
    
    if place[turn - 1] == 'X' or place [turn-1] == 'O':
        print("You cannot do that! Try again")
    
    if p1 :
        place[turn - 1] = "X"
    else :
        place[turn - 1] = "O"
    p1 = not p1
    if place[0] == "X" and place[1] == "X" and place[2] == "X": #top row win X
        win = True
        grid()
    elif place[0] == "O" and place[1] == "O" and place[2] == "O": #top row win O
        win = True
        grid()
    elif place[3] == "X" and place[4] == "X" and place[5] == "X": #middle row win X
        win = True
        grid()
    elif place[3] == "O" and place[4] == "O" and place[5] == "O": #middle row win O
        win = True
        grid()
    elif place[6] == "X" and place[7] == "X" and place[8] == "X": #bottom row win X
        win = True
        grid()
    elif place[6] == "O" and place[7] == "O" and place[8] == "O": #bottom row win O
        win = True
        grid()
    elif place[0] == "X" and place[3] =="X" and place[6] == "X": #left column win X
        win = True
        grid()
    elif place[0] == "O" and place[3] == "O" and place[6] == "O": #left column win O
        win = True
        grid()
    elif place[1] == "X" and place[4] == "X" and place[7] == "X": #middle column win X
        win = True
        grid()
    elif place[1] == "O" and place[4] == "O" and place[7] == "O": #middle column win O
        win = True
        grid()
    elif place[2] == "X" and place[5] == "X" and place[8] == "X": #right column win X
        win = True
        grid()
    elif place[2] == "O" and place[5] == "O" and place[8] == "O": #right column win O
        win = True
        grid()
    elif place[0] == "X" and place[4] == "X" and place[8] == "X": #diagonal 1 X
        win = True
        grid()
    elif place[0] == "O" and place[4] == "O" and place[8] == "O": #diagonal 1 O
        win = True
        grid()
    elif place[2] == "X" and place[4] == "X" and place[6] == "X": #diagonal 2 X
        win = True
        grid()
    elif place[2] == "O" and place[4] == "O" and place[6] == "O": #diagonal 2 O
        win = True
        grid()

print ("Player " + str(int(p1 + 1)) + " wins!\n")
```
